```python
import pandas as pd
import os

# handles a list of files on the filesystem with names matching a pattern
import glob
```

```python
# access files in folder "data"
os.listdir('data')
```

```python
path = "data"
filelist = glob.glob(path + "/*tripdata.csv")
filelist
```

```python
list = []
for f in filelist:
    df = (pd.read_csv(f))
    list.append(df)
```

```python
concat_list = pd.concat(list, axis = 0, ignore_index = True)
concat_list
```

```python
concat_list.count()
```

```python

concat_list.to_csv("data/tripdata.csv")
```
